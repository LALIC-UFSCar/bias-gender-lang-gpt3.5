```python
import pandas as pd
```

```python
PATH_DATA_01 = '../data/01_raw/'
PATH_DATA_02 = '../data/02_intermediate/'
```

```python
def tsv_to_csv(file_name):
    file_path = f'{PATH_DATA_01}{file_name}'
    df = pd.read_csv(file_path, sep='\t', header=None, names=['Polaridade', 'Frase'])
    return df
```

```python
dev = 'dev.tsv'
df_dev = tsv_to_csv(dev)

test = 'test.tsv'
df_test = tsv_to_csv(test)

train = 'train.tsv'
df_train = tsv_to_csv(train)

train_other = 'train_other.tsv'
df_train_other = tsv_to_csv(train_other)
```

```python
df = pd.concat([df_dev, df_test, df_train, df_train_other], ignore_index=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
file_name = 'frases.csv'
file_path = f'{PATH_DATA_02}'
df.to_csv(file_name)
```
